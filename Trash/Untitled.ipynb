```python
import pandas as pd
import glob
import os
```

```python
# Folder path where your CSV files are stored
folder_path = "Top3_Data"  # Replace with your folder path

# Get a list of all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# List to store individual DataFrames
dataframes = []

# Read each CSV file and append to the list
for file in csv_files:
    single_df = pd.read_csv(file)
    dataframes.append(single_df)

# Concatenate all DataFrames into a single DataFrame
df = pd.concat(dataframes, ignore_index=True)
df.head()
```

```python
# df = pd.read_csv('Top3_Data/DBTL6.csv')
```

```python
#Import translator
name_df = pd.read_csv('Figure_Data/proteomics_id_translator_240305.csv')  # Assuming a file with 'original' and 'locus' column
```

```python
# Step 1: Translate Protein.Group names to locus
translator_dict = {}
nontranslated = 0
for orig_label in df['Protein.Group']:
    matching_idx = name_df['extracted'].str.contains(orig_label, regex=False)
    if sum(matching_idx) == 1:
        translator_dict[orig_label] = name_df.loc[matching_idx, 'locus'].values[0]
    else:
        translator_dict[orig_label] = orig_label  # Keep the original name if no match
        nontranslated += 1
print(f"In total, N = {nontranslated}/{len(df['Protein.Group'])} proteins were not translated to locus names")

# Apply the translation to both files
df['Protein.Group'] = df['Protein.Group'].map(translator_dict)
df.tail()
```

```python
df2=df.copy()
filtered_df = df2.dropna(subset=[col for col in df2.columns if col.startswith('Protein.Group')], how='all')
filtered_df.head()
```

```python
# Step 1: Clean 'Sample' column by removing suffixes (_P1, _bat2)
filtered_df['Cleaned_Column'] = filtered_df['Sample'].str.replace(r'(_P\d+|_Bat\d+)$', '', regex=True)
split_columns = filtered_df['Cleaned_Column'].str.split('_PP_', expand=True)

# Step 3: Rename resulting split columns for clarity
split_columns.columns = [f"Split_{i+1}" for i in range(split_columns.shape[1])]
for col in split_columns.columns:
    split_columns[col] = split_columns[col].apply(lambda x: f"PP_{x}" if pd.notna(x) and not x.startswith('PP_') and x != 'Control' else x)

# Step 5: Handle 'Control' rows to replace split parts with 'Control'
split_columns = split_columns.where(~filtered_df['Sample'].str.contains('Control'), 'Control')

# Step 6: Combine split columns back with the filtered DataFrame
result = pd.concat([filtered_df, split_columns], axis=1)

# Display the last few rows of the result
result.tail()
```

## Seems good to here

```python
import pandas as pd
from scipy.stats import ttest_ind

result2 = result.copy()

# Step 1: Define a function to check for PP_0815 in any split column
def contains_PP_1317(row):
    return any(row[col] == 'PP_1317' for col in result2.columns if col.startswith('Split_'))

# Step 2: Apply the function and create a new boolean column
result2['Contains_PP_1317'] = result2.apply(contains_PP_1317, axis=1)

# Step 3: Perform t-test for each unique Protein.Group
results = []

for protein in result2['Protein.Group'].unique():
    # Subset data for the current protein group
    subset = result2[result2['Protein.Group'] == protein]
    
    # Split into two groups: those with and without PP_0815
    group_with_PP_1317 = subset[subset['Contains_PP_1317']]['%_of protein_abundance_Top3-method']
    group_without_PP_1317 = subset[~subset['Contains_PP_1317']]['%_of protein_abundance_Top3-method']
    
    # Perform t-test if both groups have at least 2 values
    if len(group_with_PP_1317) > 1 and len(group_without_PP_1317) > 1:
        t_stat, p_val = ttest_ind(group_with_PP_1317, group_without_PP_1317, equal_var=False)
    else:
        t_stat, p_val = None, None  # Not enough data for t-test
    
    # Append results
    results.append({
        'Protein.Group': protein,
        'Group_with_PP_1317_Mean': group_with_PP_1317.mean() if not group_with_PP_1317.empty else None,
        'Group_without_PP_1317_Mean': group_without_PP_1317.mean() if not group_without_PP_1317.empty else None,
        'T-Statistic': t_stat,
        'P-Value': p_val
    })

# Step 4: Convert results to a DataFrame
results_df = pd.DataFrame(results)
results_df.to_csv('results_df_PP_1317.csv', index=False)

# Display the results
print(results_df)
```
