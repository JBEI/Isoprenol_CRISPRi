- EDD study slug

**Files generated by running this notebook:**
 
- a pickle file containing a trained ART object

Keep the strains that contain the defined targets:

```python
def find_strains_with_targets(df, targets):
    strains = df_s["Strain"].unique()
    keep_strains = []
    for strain in strains:
        if strain[3:7] in targets:
            keep_strains.append(strain)
            
    keep_strains.sort()
            
    return keep_strains
            
```

```python
keep_strains = find_strains_with_targets(df_s, targets)
```

```python
df_s = df_s[df_s['Strain'].isin(keep_strains)]
```

### Find recommendations

Alternatively, you can change those parameters to generate recommendations from a longer/shorter sampling run or a different exploration-exploitation strategy:

```python
art.niter = 100000
art.rel_rec_distance = 1.8
art.optimize()
```

Sampling-based optimization and generation of recommendations was ran above with the default options (`niter=100000` for number of samples collected, `alpha=None` for a purely exploitative strategy) and `rel_rec_distance=40%` for relative recommendations distance.


Calculate and visualize the success probabilities if the recommended strains are being engineered.

To inspect the global importance of each feature in the model for all instances we run:

```python
shap.summary_plot(shap_values_e)
```

This bar plot shows that Class B acid phosphatase, Hydroxymethylglutaryl-CoA synthase and Acetyl-CoA acetyltransferase/HMG-CoA reductase are the top three factors driving the model's prediction of Isoprenol

```python
shap.plots.heatmap(shap_values_e, instance_order=shap_values_e.sum(1), max_display=15)
```

```python
clustering = shap.utils.hclust(X_df, art_scaleinput.y) # by default this trains (X.shape[1] choose 2) 2-feature XGBoost models
shap.plots.bar(shap_values_e, clustering=clustering)
```

```python
# use Kernel SHAP to explain the ART's predictions
explainer = shap.KernelExplainer(model=f, data=X_df)
shap_values = explainer.shap_values(X=X_df, nsamples=10) # runs ART 10 times
```
