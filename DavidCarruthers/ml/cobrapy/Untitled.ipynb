```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('strain_predictions.csv', index_col=0)
df['c'] = df['strain'].apply(lambda x: int(x[-1]))
df['c'].unique()
df['is_control'] = ['ontrol' in x for x in df['strain']]
```

```python
fig, ax = plt.subplots()
sns.scatterplot(data = df, 
                x = 'normalized_isoprenol_prediction',
                y = 'measured_isoprenol',
                hue = 'is_control'
               )

ax.set_title('GSM predictions for C0-C3')
ax.set_xlabel('Normalized Flux Prediction')
ax.set_ylabel('Normalized Measured Isoprenol')
```
