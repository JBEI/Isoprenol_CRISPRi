`data_df_3` is a subset of `data_df_013`.

```python
data_df_013 = data_df_013[data_df_013['pass_to_art']]
```

### Compare titer distributions

```python
fig, ax = plt.subplots(figsize = (5,5))

sns.violinplot(data = data_df_013, y = 'isoprenol', x = 'cycle', hue = 'is_control', cut = 0, width=1.1)
sns.move_legend(ax, "upper left")
```

```python
all_sorted_gene_lists = data_df_013.loc[data_df_013['cycle'].isin([1,3]), 'sorted_gene_list'].values
cycle3_sorted_gene_lists = data_df_013.loc[data_df_013['cycle'] == 3, 'sorted_gene_list'].values
all_genes = [gene for sgl in all_sorted_gene_lists for gene in sgl]
cycle3_genes = [gene for sgl in cycle3_sorted_gene_lists for gene in sgl]
print(f"There are N = {len(np.unique(all_genes))} unique genes in dbtl0/1/3 and N = {len(np.unique(cycle3_genes))} unique genes in dbtl3only.")
```

```python
print(f"DBTL0/1/3 mean titer: {data_df_013['isoprenol'].mean()}\t max titer: {data_df_013['isoprenol'].max()}\t control titer: {data_df_013.loc[data_df_013['is_control'],'isoprenol'].mean()}")
print(f"DBTL3only mean titer: {data_df_013.loc[data_df_013['cycle'] == 3, 'isoprenol'].mean()}\t max titer: {data_df_013.loc[data_df_013['cycle'] == 3, 'isoprenol'].max()}\t control titer: {data_df_013.loc[((data_df_013['cycle'] == 3) & (data_df_013['is_control'])), 'isoprenol'].mean()}")
```

```python
print(f"DBTL0/1/3 training points: {len(data_df_013)}")
print(f"DBTL3only training points: {len(data_df_013.loc[data_df_013['cycle'] == 3, :])}")
```
