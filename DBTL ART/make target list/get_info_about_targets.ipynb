```python
import cobra
model = cobra.io.load_json_model('../ml/cobrapy/data/iJN1463_IPP_bypass.json')
```

```python
#Update one gene ID
cobra.manipulation.modify.rename_genes(model = model, rename_dict = {'PP_s0001': 'PP_0001'})
# model.genes.get_by_id('PP_s0001').id = 'PP_0001'
print(model.genes.get_by_id('PP_0001'))
gene_ids = [g.id for g in model.genes]
```

```python
def get_gsm_reactions(x, gene_ids):
    if x in gene_ids:
        return [r.id for r in model.genes.get_by_id(x).reactions]
    else:
        return None
def get_gsm_rxn_subsystems(x, model):
    subsystem_list = []
    if x is not None:
        for rxn_id in x:
            rxn = model.reactions.get_by_id(rxn_id)
            if rxn.subsystem is not None:
                subsystem_list.append(rxn.subsystem)
    return list(set(subsystem_list))
```

```python
model.reactions.get_by_id('ATPS4rpp')
```

```python
df_merge['in_gsm'] = df_merge['pp'].apply(lambda x: x in gene_ids)
df_merge['gsm_rxns'] = df_merge['pp'].apply(lambda x: get_gsm_reactions(x, gene_ids))
df_merge['gsm_subsystems'] = df_merge['gsm_rxns'].apply(lambda x: get_gsm_rxn_subsystems(x, model))
df_merge['num_rxns'] = df_merge['gsm_rxns'].apply(lambda x: len(x) if x is not None else 0)
df_merge['num_subsystems'] = df_merge['gsm_subsystems'].apply(lambda x: len(x) if x is not None else 0)
df_merge['subsystem1'] = df_merge['gsm_subsystems'].apply(lambda x: x[0] if len(x)>0 else '-')
df_merge['in_gsm'].sum()
```

# Make target list
Here, I make a csv file containing information about the initial targets we chose:
- Locus name (PP_XXXX) - done
- Protein name - From uniprot - done
- Protein function - From uniprot - done
- Source (heuristic vs. Flux-RETAP) - From Ian's docs - done
- RbTnSeq essentiallity
- Isoprenol titer - From DBTL0 results - done
- dCas9 expression normalized to control - from DBTL0 results - done
- Whether it passed filters - From DBTL0 results - done
- Whether it was included after DBTL3 - From DBTL3 results
- Anything else? 

```python
columns_to_keep = [
    'pp',
    'Protein names',
    'retap_target',
    'in_rbtnseq',
    'isoprenol_mean', 
    'dCas9_expr',
    'target_control_ratio',
    'pass_dbtl0_filter',
    'pass_dbtl3'
]
df_table = df_merge[columns_to_keep].copy()

df_table['retap_target'] = df_table['retap_target'].apply(lambda x: 'FR' if x else 'H')
df_table['in_rbtnseq'] = df_table['in_rbtnseq'].apply(lambda x: 'N' if x else 'Y') 
df_table['isoprenol_mean'] = np.round(df_table['isoprenol_mean'], 2)
df_table['dCas9_expr'] = np.round(df_table['dCas9_expr'], 2)
df_table['target_control_ratio'] = np.round(df_table['target_control_ratio'], 2)
rename_dict = {
    'pp': 'Locus number',
    'Protein names': 'Protein name',
    'retap_target': 'Source (Flux-RETAP or Heuristic)',
    'isoprenol_mean': 'Mean isoprenol titer (mg/L)',
    'dCas9_expr': 'dCas9/control',
    'target_control_ratio': 'Target/control',
    'pass_dbtl0_filter': 'Passed DBTL0 filter?',
    'pass_dbtl3': 'Used after DBTL2/3?',
    'in_rbtnseq': 'RbTn-Seq Essential?'
}

df_table = df_table.rename(columns = rename_dict)

df_table.to_csv('Target_table.csv', index=None)
```
