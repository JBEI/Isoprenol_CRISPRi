```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the first .csv with letter codes
data = pd.read_csv('241117_COGs.csv')

# Extract the 'Cat' column
cat_data = data['Cat']

# Parse rows with multiple letters
letters = cat_data.str.split().explode()

# Count the frequency of each letter
letter_counts = letters.value_counts()

# Load the second .csv with mappings (Letter → Function + Description)
# Replace with your actual mapping file and columns
mapping_data = pd.DataFrame({
    "Letter": ['A', 'B', 'C', 'D'],
    "Function": ['Cellular process', 'Metabolism', 'DNA repair', 'Poorly characterized'],
    "Description": ['Cellular processes and signaling', 'Metabolism', 'Information storage and processing', 'Poorly characterized']
})

# Create a mapping dictionary
mapping_dict = dict(zip(mapping_data['Letter'], mapping_data['Function']))
description_dict = dict(zip(mapping_data['Function'], mapping_data['Description']))

# Map letters to functions
letter_counts.index = letter_counts.index.map(mapping_dict)

# Group by descriptions
letter_counts_with_description = pd.DataFrame({'Function': letter_counts.index, 'Frequency': letter_counts.values})
letter_counts_with_description['Description'] = letter_counts_with_description['Function'].map(description_dict)

# Group by the higher-level descriptions and sum frequencies
grouped = letter_counts_with_description.groupby(['Description', 'Function'])['Frequency'].sum().unstack()

# Plot
plt.figure(figsize=(10, 6))
grouped.plot(kind='barh', stacked=True, figsize=(10, 8), cmap='tab20', edgecolor='black')

plt.title('Functions Grouped by Higher-Level Descriptions', fontsize=14)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Descriptions', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the first .csv with letter codes
data = pd.read_csv('241117_COGs.csv')

# Extract the 'Cat' column
cat_data = data['Letter']

# Parse rows with multiple letters
letters = cat_data.str.split().explode()

# Count the frequency of each letter
letter_counts = letters.value_counts()

# Sort by index (alphabetically by letters)
letter_counts_sorted = letter_counts.sort_index()

# Load the second .csv with letter-to-function mappings
mapping_data = pd.read_csv('241117_Codex.csv')  # Replace with actual filename

# Create a mapping dictionary (assuming columns 'Letter' and 'Function')
mapping_dict = dict(zip(mapping_data['Letter'], mapping_data['Function']))

# Map letters to functions
letter_counts_sorted.index = letter_counts_sorted.index.map(mapping_dict)

# Plot the horizontal bar plot with functions
plt.figure(figsize=(6, 6))
letter_counts_sorted.plot(kind='barh', edgecolor='black', color='skyblue')
plt.title('Function Incidence', fontsize=14)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Functions', fontsize=12)
plt.tight_layout()  # Ensures the layout fits well
plt.show()
```

```python
# Load the first .csv with letter codes
data = pd.read_csv('241117_COGs2.csv')

# Extract the 'Cat' column
cat_data = data['Letter']
letters = cat_data.str.split().explode()
letter_counts = letters.value_counts()

# Load the second .csv with letter-to-function mappings
mapping_data = pd.read_csv('241117_Codex.csv')  # Replace with actual filename

# Create a mapping dictionary (assuming columns 'Letter' and 'Function')
mapping_dict = dict(zip(mapping_data['Letter'], mapping_data['Function']))

# Map letters to functions
letter_counts.index = letter_counts.index.map(mapping_dict)

# Now, let's consider the 'Category' column in the original data, which bifurcates into "A" or "B"
# Add the Category data from the original CSV to the letters dataframe
source_data = data['Source'].repeat(cat_data.str.split().str.len())  # Repeat the category per letter in the row

# Combine the letters with the corresponding categories
letters_df = pd.DataFrame({
    'Letter': letters,
    'Source': source_data
})

# Count occurrences of Letter-Category pairs
letter_category_counts = letters_df.groupby(['Letter', 'Source']).size().unstack(fill_value=0)

# Map the letter codes to functions
letter_category_counts.index = letter_category_counts.index.map(mapping_dict)

# Plot a stacked bar plot
plt.figure(figsize=(8, 6))
letter_category_counts.plot(kind='barh', stacked=True, edgecolor='black', color=['skyblue', 'salmon'])
plt.title('Function Incidence by Category')
plt.legend()
plt.xlabel('Frequency')
plt.ylabel('Functions')
plt.tight_layout()  # Ensures the layout fits well
plt.show()
```
