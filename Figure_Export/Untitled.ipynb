```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate some example data
np.random.seed(42)
X = np.random.rand(100) * 10  # Independent variable (100 random points)
y = 3 * X + 7 + np.random.normal(0, 3, 100)  # Dependent variable with noise

# Add a constant term for the intercept
X_with_constant = sm.add_constant(X)

# Perform linear regression
model = sm.OLS(y, X_with_constant)
results = model.fit()

# Print the summary of the regression model
print(results.summary())

# Get the residuals from the regression model
residuals = results.resid

# Plotting the Q-Q plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
sm.qqplot(residuals, line='45', ax=ax, fit=True)
ax.set_title('Q-Q Plot of Residuals')
plt.show()

```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate example data
np.random.seed(42)
X = np.random.rand(100) * 10  # Independent variable
y = 3 * X + 7 + np.random.normal(0, 3, 100)  # Dependent variable with noise

# Add a constant (intercept term)
X_with_constant = sm.add_constant(X)

# Fit linear regression model
model = sm.OLS(y, X_with_constant)
results = model.fit()

# Influence plot
fig, ax = plt.subplots(figsize=(8, 6))
sm.graphics.influence_plot(results, ax=ax, criterion="cooks")
plt.title('Influence Plot')
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate example data
np.random.seed(42)
X = np.random.rand(100) * 10  # Independent variable
y = 3 * X + 7 + np.random.normal(0, 3, 100)  # Dependent variable with noise

# Add a constant (intercept term)
X_with_constant = sm.add_constant(X)

# Fit linear regression model
model = sm.OLS(y, X_with_constant)
results = model.fit()

# Get fitted values and residuals
fitted_values = results.fittedvalues
residuals = results.resid

# Residuals vs Fitted plot
plt.figure(figsize=(8, 6))
plt.scatter(fitted_values, residuals)
plt.axhline(y=0, color='red', linestyle='--')  # Add a horizontal line at 0
plt.title('Residuals vs Fitted Plot')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```
