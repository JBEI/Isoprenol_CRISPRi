# Analysis of the proteomics data of isoprenol producing strains with single and multiple gene targets



We will make a final choice of the targets by looking into:

- repression levels relative to the control and 
- repression levels relative to the mean values of all samples excluding the one with the target
- Z-score of the above
- isoprenol levels 
- isoprenol CV
- taking into account replicate variability
- subsystems of the genes 
- operon group
- was dCas9 expressed? include only those that had it expressed

From Ian and Taek Soon: 

"We will also probably need to take into consideration the expression level of isoprenol pathway proteins (MvaE, MvaS, PMDHKQ , MK, and AphA) when selecting the target gene candidates. The expression of those proteins varies between strains. MvaE protein was not detected in all the samples."

- We look into genes that we repressed by at least 50% (there are 68 of those)
- From those we choose all the targets that Hector proposed (13 in total)
- We choose additional 7 from the rest, by looking into above criteria

Tested using **ART-DEV 3.9** kernel on skynet.ese.lbl.gov.

## Helping functions

```python
def create_strain_names(df: pd.DataFrame):
    "Creates strain names by reading Line Names including replicate number"
    strains = [
        line_name.split('-')[0] for line_name in df["Line Name"].to_list()
    ]
    return strains
```

```python
def find_strains_with_targets(df, targets):
    
    strains = df["Strain"].unique()
    strains = [strain.split('_') for strain in strains]
    keep_strains = []
    
    for strain in strains:
        
        if len(strain) == 1:
            continue
        
        num_genes = len(strain) - 1
        
        # Maybe needs to be removed
        # if num_genes == 0:
            # continue
            
        count_targets = 0
        for gene in strain[1:]:
            if gene in targets:
                count_targets += 1
                
        if count_targets == num_genes:
            keep_strains.append('_'.join(strain))
            
    # Include the control strains
    keep_strains.append('Control')

    keep_strains.sort()
        
    return keep_strains
            
```
