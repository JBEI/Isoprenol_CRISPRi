```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from textwrap import wrap
from matplotlib.font_manager import FontProperties
import re
from scipy import stats
from matplotlib.lines import Line2D 

line_size=0.5
custom_params = {
    'axes.facecolor': 'white',
    'axes.edgecolor': 'black',
    'axes.grid': False,
    'axes.axisbelow': 'line',
    'axes.labelcolor': 'black',
    'figure.facecolor': 'white',
    "axes.spines.right": True, 
    "axes.spines.bottom": True,
    'xtick.direction': 'inout',
    'ytick.direction': 'out',
    'xtick.bottom': True,
    'xtick.minor.visible': False,
    'xtick.major.bottom':  True,
    'xtick.minor.bottom':  True,
    'xtick.major.pad':     7,     # distance to major tick label in points#
    'xtick.minor.pad':     3.4,     # distance to the minor tick label in points#
    'xtick.major.size':    10,
    'ytick.major.size':    4,
    'xtick.major.width':   line_size,
    'ytick.major.width':   line_size,
    'axes.linewidth': line_size,
    'xtick.minor.size':    2,
    'pdf.fonttype': 42}

sns.set_theme(style="ticks", rc=custom_params, font="Arial",font_scale=0.7)

enmax_palette = ["#648FFF", "#FE6100", "#785EF0", "#DC267F", "#FFB000"]
sns.set_palette(palette=enmax_palette)
```

```python
#Target_vs_NonTarget
figname = "Target_vs_NonTarget"
df = pd.read_csv('Figure_Data\\241003_KO_vs_KD.csv')
data=df.copy()
sns.set_palette(palette=enmax_palette)
w=len(data)/36
print(w)
plt.figure(figsize=(w,2.5))

df_filtered = df[(df['Screen'] == 'KO') & (df['Guide'].isin(['Target', 'Non-Target']))]

# Step 2: Group by Line Name and Guide, calculate mean and stdev (standard deviation)
df_agg = df_filtered.groupby(['Line Name', 'Guide']).agg(
    Value_mean=('Value', 'mean'),
    Value_std=('Value', 'std')
).reset_index()

# Step 2: Pivot the DataFrame to get separate columns for Target and Non-target
df_pivot = df_agg.pivot(index='Line Name', columns='Guide')

# Flatten the multi-index columns
df_pivot.columns = ['_'.join(col).strip() for col in df_pivot.columns.values]

# Step 3: Drop rows where either Target or Non-target is missing (NaN)
df_pivot = df_pivot.dropna(subset=['Value_mean_Target', 'Value_mean_Non-Target'])

# Step 4: Calculate the R^2 value using the cleaned DataFrame
slope, intercept, r_value, p_value, std_err = stats.linregress(df_pivot['Value_mean_Target'], df_pivot['Value_mean_Non-Target'])
r_squared = r_value ** 2

x_values = range(0,500)
y_values = slope * x_values + intercept  # Linear regression line equation

# Step 5: Plot with error bars
plt.errorbar(df_pivot['Value_mean_Target'], df_pivot['Value_mean_Non-Target'],
             xerr=df_pivot['Value_std_Target'], yerr=df_pivot['Value_std_Non-Target'],
             fmt='o', color='#648FFF', ecolor='black', capsize=3, capthick=line_size,
             markeredgecolor='black', markeredgewidth=line_size, linewidth=line_size)

# Add y = x line
plt.plot([0, 500],
         [0, 500], 'k-', linewidth = line_size)

plt.plot(x_values, y_values, 'r--', label=f'Correlation: y={slope:.2f}x + {intercept:.2f} (R²={r_squared:.3f})', linewidth=line_size)

# Labels and title
plt.title(f'KO Strains with Target or Non-Target gRNA', weight = 'bold', fontsize=8)

legend_line = Line2D([0], [0], color='black', linestyle='-', linewidth=line_size, label='Target = Non-Target')
regression_line = Line2D([0], [0], color='red', linestyle='--', linewidth=line_size, label=f'R²={r_squared:.3f}')

# Update the legend to include both lines
plt.legend(handles=[legend_line, regression_line], loc='upper left')

plt.xlabel('KO Strain with Control guide')
plt.ylabel('KO Strain with PP_#### guide')
plt.xlim(0,425)
plt.ylim(0,425)


plt.savefig(f'Figure_Export\\{figname}.svg',dpi=1200, bbox_inches = 'tight')
plt.show()
```

```python
#KO_vs_KD
figname = "KO_vs_KD"
df = pd.read_csv('Figure_Data\\241003_KO_vs_KD.csv')
data=df.copy()

sns.set_palette(palette=enmax_palette)
w=len(data)/36
print(w)
plt.figure(figsize=(w,2.5))

# Filter
control_lines = ['Control'] 
df_filtered = df[(df['Guide'] == 'Target') & (~df['Line Name'].isin(control_lines))]

# Aggregate to calculate mean and stdev
df_agg = df_filtered.groupby(['Line Name', 'Screen']).agg(
    Value_mean=('Value', 'mean'),
    Value_std=('Value', 'std')
).reset_index()

# Separate columns and flatten
df_pivot = df_agg.pivot(index='Line Name', columns='Screen')
df_pivot.columns = ['_'.join(col).strip() for col in df_pivot.columns.values]

# Drop NaNs
df_pivot = df_pivot.dropna(subset=['Value_mean_KO', 'Value_mean_dCas9'])
df_pivot

# Calculate the R^2 value
slope, intercept, r_value, p_value, std_err = stats.linregress(df_pivot['Value_mean_KO'], df_pivot['Value_mean_dCas9'])
r_squared = r_value ** 2

x_values = range(0,500)
y_values = slope * x_values + intercept  # Linear regression line equation

plt.errorbar(df_pivot['Value_mean_KO'], df_pivot['Value_mean_dCas9'],
             xerr=df_pivot['Value_std_KO'], yerr=df_pivot['Value_std_dCas9'],
             fmt='o', color='#648FFF', ecolor='black', capsize=3, capthick=line_size,
             markeredgecolor='black', markeredgewidth=line_size, linewidth=line_size)

#Plotting linear regression line
plt.plot(x_values, y_values, 'r--', label=f'Correlation: y={slope:.2f}x + {intercept:.2f} (R²={r_squared:.3f})', linewidth=line_size)

# Add y = x line
plt.plot([0, 500],
         [0, 500], 'k-', linewidth=line_size)

# Labels and title
plt.title(f'Knockout compared to CRISPRi Strains', weight = 'bold', fontsize=8)
legend_line = Line2D([0], [0], color='black', linestyle='-', linewidth=line_size, label='Knockout = CRISPRi')
regression_line = Line2D([0], [0], color='red', linestyle='--', linewidth=line_size, label=f'R²={r_squared:.3f}')

plt.legend(handles=[legend_line, regression_line], loc='upper left')
plt.xlabel('CRISPRi Strain performance')
plt.ylabel('KO Strain performance')
plt.xlim(0,425)
plt.ylim(0,425)

plt.savefig(f'Figure_Export\\{figname}.svg',dpi=1200, bbox_inches = 'tight')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator

# Load and copy the data
figname = "Arrays_Sum"
df = pd.read_csv('Figure_Data\\241002_CRISPRi_Total_Arrays.csv')
data = df.copy()

# Set the palette
enmax_palette2 = ["#FFB000", "#DC267F"]
sns.set_palette(palette=enmax_palette2)

# Get unique categories
arrays = data['Array'].unique()
statuses = data['Status'].unique()
cycles = data['Cycle'].unique()

# Sort the cycles to ensure they are in order
cycles = np.sort(cycles)

# Create a pivot table with multi-index columns (Array and Status)
pivot_table = data.pivot_table(index='Cycle', columns=['Array', 'Status'], values='# Arrays', aggfunc='sum', fill_value=0)

# Calculate the width based on the number of cycles
w = len(cycles)
print(w)

# Set up the figure
fig, ax = plt.subplots(figsize=(w, 2.5))

# Define bar width and positions
total_bar_width = 0.8  # Total width for all bars in a group
n_arrays = len(arrays)
bar_width = total_bar_width / n_arrays
bar_positions = np.arange(len(cycles))

# Create a color palette for statuses
status_colors = sns.color_palette(n_colors=len(statuses))
status_color_dict = dict(zip(statuses, status_colors))

# Plot bars for each array
for i, array in enumerate(arrays):
    # Position of the group of bars
    x_positions = bar_positions + (i - n_arrays / 2) * bar_width + bar_width / 2
    # Bottom positions for stacking
    bottoms = np.zeros(len(cycles))
    # Plot each status on top of the previous
    for status in statuses:
        # Extract data for this array and status
        values = []
        for cycle in cycles:
            # Safely get the value or 0 if it doesn't exist
            try:
                value = pivot_table.loc[cycle, (array, status)]
            except KeyError:
                value = 0
            values.append(value)
        # Plot the bar segment
        ax.bar(x_positions, values, bar_width, bottom=bottoms, color=status_color_dict[status], edgecolor="0", linewidth=line_size, label=status if i == 0 else "")
        # Update bottoms for stacking
        bottoms += values

# Set x-axis labels and ticks
ax.set_xticks(bar_positions)
ax.set_xticklabels(cycles, rotation='horizontal', style='italic')

# Add minor ticks for arrays
ax.minorticks_on()  # Enable minor ticks
minor_ticks = []
minor_labels = []

# Set minor ticks between the major ticks
for i, cycle in enumerate(cycles):
    for j, array in enumerate(arrays):
        minor_ticks.append(bar_positions[i] + (j - (n_arrays / 2)) * bar_width + bar_width / 2)
        minor_labels.append(array)

ax.set_xticks(minor_ticks, minor=True)  # Set minor tick positions
ax.set_xticklabels(minor_labels, minor=True, rotation=90, ha='center')  # Set minor tick labels

# Set labels

ax.set_ylabel('# of plasmids', labelpad=10)
ax.tick_params(axis='x', which='major', pad=40)  # Increase pad for major ticks
ax.yaxis.set_minor_locator(AutoMinorLocator(1))  # Disable minor ticks on y-axis
ax.set_xlabel('Cycle', labelpad=10)

# Remove duplicate labels in the legend
handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), title='', bbox_to_anchor=(1.05, 1), loc='upper left')

# Tighten layout and show the plot
plt.tight_layout()
plt.savefig(f'Figure_Export/{figname}.svg', dpi=1200, bbox_inches='tight', transparent=True)
plt.show()
```
